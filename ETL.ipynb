```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
import requests
```

```python
#Store csv to Data Frame
csv_file = "Resources/degrees-that-pay-back.csv"
degrees_df = pd.read_csv(csv_file)
degrees_df

new_degrees_df = degrees_df[["Undergraduate Major", "Mid-Career Median Salary",
                            "Mid-Career 25th Percentile Salary", 
                             "Mid-Career 75th Percentile Salary"]].copy()
new_degrees_df = new_degrees_df.rename(columns = {"Undergraduate Major": "major",
                                "Mid-Career Median Salary": "median_salary",
                                "Mid-Career 25th Percentile Salary": "first_percentile",
                                "Mid-Career 75th Percentile Salary": "third_percentile"})

new_degrees_df.head()
```

```python
#Clean Data
new_degrees_df['median_salary'] = [x.strip('$') for x in new_degrees_df['median_salary']]
new_degrees_df['first_percentile'] = [x.strip('$') for x in new_degrees_df['first_percentile']]
new_degrees_df['third_percentile'] = [x.strip('$') for x in new_degrees_df['third_percentile']]

new_degrees_df['median_salary']=new_degrees_df['median_salary'].str.replace(',','')
new_degrees_df['first_percentile']=new_degrees_df['first_percentile'].str.replace(',','')
new_degrees_df['third_percentile']=new_degrees_df['third_percentile'].str.replace(',','')
new_degrees_df.head()
```
